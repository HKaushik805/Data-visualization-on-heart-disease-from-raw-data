```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

## Distribution Plot (distplot) for Numeric Column

```python
sns.displot(df['Age'], kde=True)
```

```python
sns.distplot(df['Age'], hist=True)
```

```python

sns.distplot(df['Age'], hist=True, rug=True)
```

```python
sns.distplot(df['Age'], bins=10)
```

```python
sns.distplot(df['Age'], kde=True, color='blue') 
```

```python
sns.distplot(df['RestingBP'], kde=True,color='blue')
```

```python
sns.displot(df['Cholesterol'], kde=True)
```

```python
sns.distplot(df['MaxHR'], kde=True)
```

```python
sns.distplot(df['Oldpeak'], kde=True)
```

## Pie Charts for Catagorical Columns

```python
df.groupby('Sex').size().plot(kind='pie', autopct='%.1f')
```

```python
df.groupby('ChestPainType').size().plot(kind='pie', autopct='%.1f')
```

```python
df.groupby('RestingECG').size().plot(kind='pie', autopct='%.1f')
```

```python
df.groupby('ST_Slope').size().plot(kind='pie', autopct='%.1f')
```

```python
df.groupby('HeartDisease').size().plot(kind='pie', autopct='%.1f')
```

## Violin Plot

```python
sns.violinplot(df['Age'])
```

```python
sns.violinplot(y=df['Sex'], x=df['HeartDisease'])
```

```python
sns.violinplot(y=df['Age'], x=df['HeartDisease'])
```

```python
sns.violinplot(y=df['Cholesterol'], x=df['HeartDisease'])
```

## JoinPlot

```python
sns.jointplot(x='Age', y='MaxHR', data = df)
```

```python
sns.jointplot(x='Oldpeak', y='HeartDisease', data = df, kind='reg')
```

```python
sns.jointplot(x='HeartDisease', y='MaxHR', data = df, kind='reg')
```
